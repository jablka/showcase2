```python
import matplotlib.pyplot as plt

img = plt.imread('flower.jpg') # I changed from png to jpg, to save some space...
img = img.copy()  # make img writable
plt.imshow(img)
```

challenge:

```python
# Draw a blue square around the flower
# Top-left: 190x350
# Bottom-right: 680x850
# Line width: 5
```

solution:

```python
x1 , y1 = 190, 350
x2 , y2 = 680, 850

width = 5
color = [0, 0, 0xFF] # blue

img[y1:y1+width, x1:x2 ] = color
img[y2:y2+width, x1:x2 ] = color

img[y1:y2, x1:x1+width] = color
img[y1:y2, x2:x2+width] = color

plt.imshow(img)
```

```python
# first dimension is rows (y)
tl_x, tl_y = 350, 190
br_x, br_y = 850, 680
width = 5

color = [0, 0, 0xFF] # blue

# Top line
img[tl_x:tl_x+width, tl_y:br_y] = color
# Bottom line
img[br_x:br_x+width, tl_y:br_y] = color
# Left line
img[tl_x:br_x, tl_y:tl_y+width] = color
# Right line
img[tl_x:br_x, br_y-width:br_y] = color

plt.imshow(img)
```
