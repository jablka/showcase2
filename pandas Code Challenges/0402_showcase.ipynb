## Conditional filtering

```python
import pandas as pd
```

```python
df = pd.read_csv("./data/marketing_campaign.csv", sep='\t')
df.head()
```

Task:  
Extract customers without a partner, who are born after 1990.

```python
# let's see what values occur in the column
df['Marital_Status'].value_counts()
```

```python
# creating a mask
mask = (df['Marital_Status'] != 'Married') & (df['Marital_Status'] != 'Together')
mask
```

```python
# second mask
mask2 = df['Year_Birth'] > 1990
mask2
```

```python
# result
df.loc[ mask & mask2 , :]
```

```python
# done
```

```python
# NOTE: author provided this solution, which gives the same results:
df.loc[(~df['Marital_Status'].isin(['Married', 'Together'])) & (df['Year_Birth'] > 1990)]
```
