## Grouping data

```python
import pandas as pd
```

```python
df = pd.read_csv("./data/marketing_campaign.csv", sep='\t')
df.info()
```

```python
# Task:
# Find the average amount of purchased products based on customers' marital status:
```

```python
mylist= ["MntWines",        
"MntFruits",       
"MntMeatProducts", 
"MntFishProducts", 
"MntSweetProducts",
"MntGoldProds"]    

mylist
```

```python
df.groupby('Marital_Status')[mylist].mean()
```

```python
# done
```

```python
df.groupby('Marital_Status').mean(numeric_only=True) # doing mean of every column with numeric values
```

```python
df.groupby('Marital_Status').mean(numeric_only=True)[mylist] # and showing only columns we're interested in.
```
