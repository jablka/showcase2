### Create new columns from existing ones  
(three solutions)

```python
# Task: 
# Create new columns for displacement-power ratio, weight per cylinder, acceleration per unit power
```

```python
# displacement / horsepower
# weight / cylinders
# acceleration / horsepower
```

```python
import pandas as pd
```

```python
df = pd.read_csv("./data/auto_mpg.csv",
                 header=0,
                 names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower','Weight',\
                          'Acceleration', 'Model Year', 'Origin'])

df
```

##### 1st solution:

```python
# first (simplest) solution:
df['d_p_h'] = df['Displacement'] / df['Horsepower']
df['w_p_c'] = df['Weight'] / df['Cylinders']
df['a_p_h'] = df['Acceleration'] / df['Horsepower']
df
```

##### 2nd solution:

```python
# solution using assign()
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.assign.html
```

```python
df = pd.read_csv("./data/auto_mpg.csv",
                 header=0,
                 names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower','Weight',\
                          'Acceleration', 'Model Year', 'Origin'])
```

```python
df = df.assign( d_p_h = df['Displacement'] / df['Horsepower'], 
                w_p_c = df['Weight'] / df['Cylinders'],
                a_p_h = df['Acceleration'] / df['Horsepower']
            ) # tu inplace nefunguje...

df
```

##### 3rd solution:

```python
# solution using insert()
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.insert.html
```

```python
df = pd.read_csv("./data/auto_mpg.csv",
                 header=0,
                 names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower','Weight',\
                          'Acceleration', 'Model Year', 'Origin'])
```

```python
df.insert( len(df.columns), column='d_p_h', value=(df['Displacement'] / df['Horsepower']) ) # is always inplace
df.insert( len(df.columns), column='w_p_c', value=(df['Weight']/df['Cylinders']) ) 
df.insert( len(df.columns), column='a_p_h', value=(df['Acceleration'] / df['Horsepower']) )
df
```
