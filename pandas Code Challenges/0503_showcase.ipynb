### Applying a custom aggregate function

```python
# Task:
# Apply a custom aggregate function 
# on the number_of_web_purchases (column 'NumWebPurchases') 
# of customers grouped by their Education level.
```

my first try (applying the function just on Series) :

```python
def myfunc(ser:pd.Series):
    return ser.sort_values(ascending=False).head()
```

```python
df.groupby('Education')['NumWebPurchases'].apply(myfunc)
```

second try (applying the function on the Dataframe) :

```python
def myfunc(dfrm):
    return dfrm.sort_values(by='NumWebPurchases' , ascending=False).head()
```

```python
df.groupby('Education').apply(myfunc)
```

```python
purchases = ['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']
df.groupby('Education').apply(myfunc)[purchases]
```

finally author's solution (more flexible in function parameters) :

```python
def top(df, n=5, column="NumWebPurchases"):
    return df.sort_values(by=column)[-n:]
```

```python
purchases = ['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']
df.groupby(['Education']).apply(top)[purchases]
```
