```python
import pandas as pd
```

```python
df = pd.read_csv('olympics.csv',skiprows=4)
df
```

### 1.
### Plot the number of medals achieved by the Chinese team (men and women) in Beijing 2008, using Matplotlib and Seaborn

```python
# 
df[ (df['Edition'] == 2008) & (df['NOC']=='CHN') ]
```

```python
# selecting only Gender column
df[ (df['Edition'] == 2008) & (df['NOC']=='CHN') ]['Gender']
```

```python
# value_counts on Gender column
df[ (df['Edition'] == 2008) & (df['NOC']=='CHN') ]['Gender'].value_counts()
```

```python
# simple plotting, directly from pandas
df[ (df['Edition'] == 2008) & (df['NOC']=='CHN') ]['Gender'].value_counts().plot(kind='bar');
```

```python
# now with seaborn
import seaborn as sns
```

```python
# selecting Edition and Country
mydata = df[ (df['Edition'] == 2008) & (df['NOC']=='CHN') ]
mydata
```

```python
# bar plot on Gender
sns.countplot(data=mydata, x='Gender', width=0.5); 
```

### 2.
### Plot the number of gold, silver, and bronze medals for each gender.

```python
# just a reminder:
# mydata is China team in 2008
# we are interested in Gender and Medal columns
mydata[['Gender','Medal']] 
```

```python
sns.countplot(data=mydata, x='Medal', hue='Gender' );
```

```python
# let's order the columns and change the color (this is in fact taken from the author's solution)
sns.countplot(data=mydata, x='Medal', hue='Gender', order=['Gold','Silver','Bronze'] , palette='bwr');
```
