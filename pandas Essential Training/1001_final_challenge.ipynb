```python
import pandas as pd
```

```python
df = pd.read_csv('olympics.csv', skiprows=4) # data from 1896 ~ 2008
df
```

#### In every Olympics, which US athlete has won the most total number of medals? 
#### Include the athlete's discipline.

```python
df[df['NOC']=='USA']
```

```python
df[df['NOC']=='USA'].groupby(['Edition','Athlete','Medal']).size()
```

```python
df2 = (
    df[df['NOC']=='USA'].groupby(['Edition','Athlete','Medal']).size().unstack(fill_value=0)
        [['Gold','Silver','Bronze']]
        .reset_index()
)
df2 # we will work with this second dataframe onwards.
```

```python
# creating a new column 'total'
df2['total'] = df2.Gold + df2.Silver + df2.Bronze 
df2 
```

```python
# sorting Edition in ascending, and 'total' column in descending order
df2 = df2.sort_values(by=['Edition', 'total'], ascending=[True,False])
df2
```

```python
# groupby Edition and show only the first athlete (the most total medals in particular Edition)
df2.groupby('Edition').first()
```

```python
# done
```

```python
# But what if there are more athletes with the same number of medals within given Edition?
# Let's find out, and show them also:
```

```python
nlargest_medalists = (
    df2.groupby('Edition')['total'].nlargest(1, keep='all')
    # using our df2 dataframe, we do
    # groupby Edition, and on column 'total' do nlargest. 
    # keep='all' means: if there are more largest value, keep them all.
)

nlargest_medalists # and the number next to the Edition is actually the index of a particular athlete
```

```python
# reseting the index
nlargest_medalists = nlargest_medalists.reset_index()
nlargest_medalists.head()
```

```python
# level_1 column actually contains indexes for our athletes
index_medalists = nlargest_medalists['level_1'].values
index_medalists
```

```python
# let's display them
df2.loc[ index_medalists ,:]
```
