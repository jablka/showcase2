```python
import matplotlib.pyplot as plt
```

```python
first_student_books = [2,4,7,3,1,5,1,0,2,3,6,4]
second_student_books = [0,5,3,1,6,4,1,1,3,4,3,2]
months = range(1,13)
```

```python
fig, ax = plt.subplots()
ax.plot(months, first_student_books, label='books first student')
ax.plot(months, second_student_books, label='books second student')
ax.set_title('Books read by students')
ax.set_ylabel('books read')
ax.set_xlabel('months')
ax.legend()
plt.show()
```
