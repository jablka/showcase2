```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# x values
df = pd.DataFrame( {'x':np.arange(-2, (8*np.pi)+1, 0.1)} ) 

# y values
df['tan']= np.tan(df['x'])
df['cotan']= 1/np.tan(df['x']) # cotangens
```

```python
df['tan'].describe() # we see here some outliers
```

```python
# cropping out the outliers
my_mask =((df['tan']>10) | (df['tan']<-10))
df.loc[my_mask, 'tan'] = np.nan
```

```python
# cropping out the outliers
my_mask =((df['cotan']>10) | (df['cotan']<-10))
df.loc[my_mask, 'cotan'] = np.nan
```

```python
my_ticks = np.arange(0, 9*np.pi, np.pi) # x_ticks ... multiples of Ï€ 
my_ticks
```

```python
plt.figure(figsize=(10,3)) # 
plt.plot(df.x, df.tan)
plt.title("tangens")

plt.ylim(-4, 4)
plt.xlim(-1, 26)

plt.yticks([0])
plt.xticks(my_ticks) # Set custom ticks and labels on the x-axis

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Remove the frame
plt.gca().spines['top'].set_visible(False)
# plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.grid(axis='x')

plt.show()
```

```python
plt.figure(figsize=(10,3)) # 
plt.plot(df.x, df.cotan)
plt.title("cotangens")
plt.ylim(-4, 4)
plt.xlim(-1, 26)

plt.yticks([0])
plt.xticks(my_ticks) # Set custom ticks and labels on the x-axis

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Remove the frame
plt.gca().spines['top'].set_visible(False)
# plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.grid(axis='x')
plt.show()
```
