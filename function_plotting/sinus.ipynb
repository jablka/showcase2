```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame( {'x':np.arange(-1, (8*np.pi)+1, 0.1)} ) # x values

df['sin']= np.sin(df['x']) # y values
```

```python
my_ticks = np.arange(0, 9*np.pi, np.pi) # x_ticks ... multiples of Ï€ 
my_ticks
```

```python
plt.plot(df.x, df.sin)

plt.yticks([-1,0,1])
plt.xticks(my_ticks) # Set custom ticks and labels on the x-axis

plt.axhline(0, color='black', linewidth=0.5) # Add a horizontal line at y=0
plt.axvline(0, color='black', linewidth=0.5) # Add a vertical line at x=0

# Remove the frame
plt.gca().spines['top'].set_visible(False)
# plt.gca().spines['bottom'].set_visible(False)
# plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.grid(True)
plt.axis('scaled')
plt.ylim(-3, 3)
plt.show()
```
